## Caso: Diagnóstico Médico con Clasificación Binaria
En este escenario, utilizaremos un conjunto de datos médicos para predecir si los pacientes tienen o no una enfermedad específica. A continuación, el código para este caso:



```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos un conjunto de datos de diagnóstico de cáncer de mama
data = load_breast_cancer()
X = data.data
y = data.target

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos y entrenamos el modelo de clasificador de vectores de soporte (SVC)
model = SVC(kernel='linear')
model.fit(X_train, y_train)

# Realizamos predicciones con el conjunto de prueba
y_pred = model.predict(X_test)

# Generamos la matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Visualizamos la matriz de confusión
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix for Disease Diagnosis')
plt.show()

# Imprimimos el informe de clasificación
print(classification_report(y_test, y_pred))
```

## Caso: Clasificación de imágenes de dígitos manuscritos
Supongamos que tenemos un modelo para clasificar imágenes de dígitos manuscritos (0-9). Utilizaremos el conjunto de datos MNIST, que es un conjunto de datos estándar en aprendizaje automático para este tipo de tareas

```python
# Vamos a utilizar el conjunto de datos MNIST disponible en sklearn y entrenar un modelo SVM para clasificar los dígitos.
# Luego, generaremos una matriz de confusión y la visualizaremos con Matplotlib.

from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Cargamos el conjunto de datos MNIST
digits = datasets.load_digits()

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.5, random_state=42)

# Creamos un clasificador: una Máquina de Vectores de Soporte
model = svm.SVC(gamma=0.001)

# Entrenamos el modelo
model.fit(X_train, y_train)

# Predecimos las etiquetas para el conjunto de prueba
y_pred = model.predict(X_test)

# Generamos la matriz de confusión
cm = metrics.confusion_matrix(y_test, y_pred)

# Visualizamos la matriz de confusión con Seaborn
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square=True, cmap='Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix', size = 15)
plt.show()
```

## Caso: Clasificación de mensajes de texto como 'spam' o 'no spam'

Supongamos que estamos construyendo un clasificador que determina si un mensaje de texto es spam. Utilizaremos un conjunto de datos ficticio para entrenar nuestro modelo y luego evaluaremos su rendimiento con una matriz de confusión.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que 'texts' es una lista de mensajes de texto y 'labels' son las etiquetas asociadas (0 para 'no spam', 1 para 'spam').
texts = ["gana dinero ahora", "reunión mañana", "oferta exclusiva", "tu factura está lista"]
labels = [1, 0, 1, 0]

# Convertimos los textos en una matriz de tokens
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.25, random_state=42)

# Entrenamos un clasificador Naive Bayes
model = MultinomialNB()
model.fit(X_train, y_train)

# Predecimos las etiquetas para el conjunto de prueba
y_pred = model.predict(X_test)

# Generamos la matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Visualizamos la matriz de confusión
sns.heatmap(cm, annot=True, fmt="d")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

# Imprimimos el informe de clasificación
print(classification_report(y_test, y_pred))
```

## Caso: Predicción de la aprobación de créditos bancarios

En este escenario, estamos interesados en construir un modelo predictivo para determinar la aprobación de créditos bancarios basándonos en factores financieros históricos de los solicitantes. Vamos a entrenar un modelo de clasificación para predecir si un crédito debe ser aprobado (positivo) o no (negativo), y luego usaremos una matriz de confusión para evaluar la precisión de nuestro modelo.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generamos un conjunto de datos sintético para el ejemplo (en la práctica, usaríamos datos reales)
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_classes=2, random_state=42)

# Dividimos el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos un clasificador de árbol de decisión
classifier = DecisionTreeClassifier(random_state=42)

# Entrenamos el clasificador
classifier.fit(X_train, y_train)

# Realizamos predicciones en el conjunto de prueba
predictions = classifier.predict(X_test)

# Generamos la matriz de confusión
cm = confusion_matrix(y_test, predictions)

# Visualizamos la matriz de confusión
sns.heatmap(cm, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
